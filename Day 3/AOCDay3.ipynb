```python
import re

def split_number_and_symbol(value):

    newvalues = []
    newvalue = ''
    row_item_symbol = ''
    char_index = -1
    chars = [*value]
    
    for char in chars:
        char_index += 1
        
        if not char.isdigit():
            row_item_symbol = char
            newvalues.append(newvalue)
            newvalue = ''
        else:
            newvalue = newvalue + char
            if char_index + 1 == len(chars):
                newvalues.append(newvalue)
            
    return [newvalues, row_item_symbol]

def parse_v2(schematic_record, row_index):
    index = -1
    row_data = []
    
    row_symbol = ''
    row_digits = []
    previous_char_type = ''
    
    for char in schematic_record:
        index += 1
        if char == '.' or char == '\n': 
            if previous_char_type == 'd':
                l = len(row_digits)
                row_data.append([row_index,index-l,l,'d',"".join(row_digits),False]) # flag as NOT adjacent

            previous_char_type = '.'
            
            
        elif char.isdigit(): 
            if not previous_char_type == 'd':
                row_digits = []
                
            row_digits.append(char)
            
            previous_char_type = 'd'
            
            
        else: #assume symbol
            if previous_char_type == 'd':
                l = len(row_digits)
                row_data.append([row_index,index-l,l,'d',"".join(row_digits),False]) # flag as NOT adjacent
                
            row_data.append([row_index,index,1,'s',char])

            previous_char_type = 's'

    # print(row_data)
    return row_data
   
            
        
        

def parse_schematic_records(schematic_records):

    parsed_data = []
    row_data = []
    row_index = -1
    for schematic_record in schematic_records:
        row_index += 1
        parsed_data = parse_v2(schematic_record, row_index)
        for row_item in parsed_data:
            row_data.append(row_item)

    return row_data

def separate_data(row_data):
    symbol_row_data = []
    number_row_data = []
    
    for row_item in row_data:
        if row_item[3] == 'd':
            number_row_data.append(row_item)
        else:
            symbol_row_data.append(row_item)
            
    return number_row_data,symbol_row_data

def evaluate_adjacency(symbol_row_data, number_row_data):
    adjacent_digits = []
    
    for symbol_item in symbol_row_data:
        for digit_item in number_row_data:
            if digit_item[5] == True:
               x = 1
            else:
                sx = symbol_item[0]
                sy = symbol_item[1]
                x1 = digit_item[0]-1
                x2 = digit_item[0]+1
                y1 = digit_item[1]-1
                y2 = digit_item[2]+digit_item[1]+1
                
                if sx in range(x1,x2) and sy in range(y1,y2):
                    digit_item[5] = True
                    
            adjacent_digits.append(digit_item)
            
        number_row_data = adjacent_digits
            
    for adjacent_item in adjacent_digits:
        print(adjacent_item)
        
def main():
    file = open('input-example.txt','r')
    schematic_records = file.readlines()

    row_data = parse_schematic_records(schematic_records)
    #for row_item in row_data:
    #   print(row_item)
        
    separated_data = separate_data(row_data)
    symbol_row_data = separated_data[1]
    number_row_data = separated_data[0]
    
    #for row_item in symbol_row_data:
    #    print(row_item)
    #print('\n')
    #for row_item in number_row_data:
    #    print(row_item)
    
    evaluate_adjacency(symbol_row_data, number_row_data)
   

# Entrypoint
main()
```
