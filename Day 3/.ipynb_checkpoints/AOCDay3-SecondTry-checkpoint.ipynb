```python
import re
import time

class AdjacentItem:
    def __init__(self,x, y1, y2, type, value):
      self.x = x
      self.y1 = y1
      self.y2 = y2
      self.type = type
      self.value = value
    def __str__(self):
      return f"{self.x},{self.y1},{self.y2},{self.type},{self.value}"

def create_dictionary(schematic_record, row_index):
    col_index = -1
    dictionary = {}
    
    for char in schematic_record:
        col_index += 1
        dictionary[row_index,col_index] = char
        
    # print(dictionary)
    return dictionary
   
def create_rows(schematic_records):
    
    row_index = -1
    rows = []
   
    for schematic_record in schematic_records:
        row_index += 1
        current_row = create_dictionary(schematic_record, row_index)

        rows.append(current_row)

    return rows

def evaluate_rows(rows):
     for row in rows:
         for key,value in row.items():
             if value not in [".","\n"] and not value.isdigit():
                 # print(key,value)
                 # identify_adjacent(key)
                 row_range = range(int(key[0])-1,int(key[0])+1)
                 col_range = range(int(key[1])-1,int(key[1])+1)
                 print(key, row_range, col_range)
                                       
                 
def identify_adjacent(key):
    # look for neighbors
   
    x = key[0]
    y = key[1]
    print(key,x,y)

def evaluate_adjacency(symbol_row_data, number_row_data):
    adjacent_digits = {}
    sum_of_parts = 0
    
    for symbol_item in symbol_row_data:
        for digit_item in number_row_data:
    
            sx = symbol_item.x
            sy = symbol_item.y1
            x1 = digit_item.x-1
            x2 = digit_item.x+1
            y1 = digit_item.y1-1
            y2 = digit_item.y2+1

            if debug == True:
                print(range(x1,x2+1), range(y1,y2+1), sx, sy, digit_item.value)
            if digit_item.value == 449:
                print(str(digit_item))
            if sx in range(x1,x2+1) and sy in range(y1,y2+1):       
                adjacent_digits[str(digit_item)] = digit_item

    for digit_item in adjacent_digits.values():
        sum_of_parts += digit_item.value

    return sum_of_parts
        
def main():
    
    start = time.perf_counter()
    
    file = open('input-example.txt','r')
    schematic_records = file.readlines()

    rows = create_rows(schematic_records)
    evaluate_rows(rows)
    
    
   
    end = time.perf_counter()
    print(f"\nExecution time {end - start:0.4f} seconds")
# Entrypoint
debug = False
main()
```
