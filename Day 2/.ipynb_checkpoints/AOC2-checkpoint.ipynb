```python
# AOC Part 1
import re
from enum import Enum

# use enum to parse the color name later
class Color(Enum):
   red = 0
   green = 1
   blue = 2

def cube_results_by_color(cubes):

    cube_results = []
    for value in Color:
        cube_results.append(0);
    
    for cube in cubes:
        # Remove whitespace 
        newcube = cube.strip().split(" ")
        cube_color = int(Color[newcube[1]].value)
        cube_results[cube_color] = int(newcube[0])

    # print("Cube Results %s" % cube_results)

    return cube_results   

def was_this_game_possible(game_record):

    valid_game = True
    max_cubes = [12,13,14] #red, green, blue

    set_results = []
    for value in Color:
        set_results.append(0)

    game_id = int(game_record.split(':')[0].replace('Game ',''))
    
    sets = game_record.split(': ')[1].split(';')
    for set in sets:
        cubes = set.split(',')
        cube_results = cube_results_by_color(cubes)

        for color in Color:
            index = color.value
            if cube_results[index] > set_results[index]:
               set_results[index] = cube_results[index]
            if cube_results[index] > max_cubes[index]:
                valid_game = False;
            
    # print("Set Results %s %i" % (set_results, too_many_of_one_color))

    result = (game_id,valid_game)
    
    return result
        

def main():
    
    results = []

    f = open('input.txt','r')
    game_records = f.read().splitlines()
        
    for game_record in game_records:
        if debug:
            print(game_record)
        result = was_this_game_possible(game_record)
        results.append(result)

    print(results)

    id_sums = 0
    for result in results:
        if result[1] == True:
            id_sums += result[0]

    print(id_sums)


# Main Program

debug = True


main()
```
