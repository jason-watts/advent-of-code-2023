```python
import re
import time

class AdjacentItem:
    def __init__(self,x, y1, y2, type, value):
      self.x = x
      self.y1 = y1
      self.y2 = y2
      self.type = type
      self.value = value
    def __str__(self):
      return f"{self.x},{self.y1},{self.y2},{self.type},{self.value}"

def parse_v2(schematic_record, row_index):
    index = -1
    row_data = []
    
    row_symbol = ''
    row_digits = []
    previous_char_type = ''
    
    for char in schematic_record:
        index += 1
        if char == '.' or char == '\n': 
            if previous_char_type == 'd':
                l = len(row_digits)
                item = AdjacentItem(row_index,index-l,index,'d',int("".join(row_digits)))
                row_data.append(item) 

            previous_char_type = '.'
             
        elif char.isdigit(): 
            if not previous_char_type == 'd':
                row_digits = []
                
            row_digits.append(char)
            previous_char_type = 'd'
            
        else: #assume symbol
            if previous_char_type == 'd':
                l = len(row_digits)
                item = AdjacentItem(row_index,index-l,index,'d',int("".join(row_digits)))
                row_data.append(item) 

            item = AdjacentItem(row_index,index,1,'s',char)
            row_data.append(item) 
  
            previous_char_type = 's'

    return row_data
   
def parse_schematic_records(schematic_records):

    parsed_data = []
    row_data = []
    row_index = -1
    for schematic_record in schematic_records:
        row_index += 1
        parsed_data = parse_v2(schematic_record, row_index)
        for row_item in parsed_data:
            row_data.append(row_item)

    return separate_data(row_data)

def separate_data(row_data):
    symbol_row_data = []
    number_row_data = []
    
    for row_item in row_data:
        if row_item.type == 'd':
            number_row_data.append(row_item)
        else:
            symbol_row_data.append(row_item)
            
    return number_row_data,symbol_row_data

def evaluate_adjacency(symbol_row_data, number_row_data):
    adjacent_digits = {}
    sum_of_parts = 0
    
    for symbol_item in symbol_row_data:
        for digit_item in number_row_data:
    
            sx = symbol_item.x
            sy = symbol_item.y1
            x1 = digit_item.x-1
            x2 = digit_item.x+1
            y1 = digit_item.y1-1
            y2 = digit_item.y2+1

            if debug == True:
                print(range(x1,x2+1), range(y1,y2+1), sx, sy, digit_item.value)
            if digit_item.value == 449:
                print(str(digit_item))
            if sx in range(x1,x2+1) and sy in range(y1,y2+1):       
                adjacent_digits[str(digit_item)] = digit_item

    for digit_item in adjacent_digits.values():
        sum_of_parts += digit_item.value

    return sum_of_parts
        
def main():
    
    start = time.perf_counter()
    
    file = open('input.txt','r')
    schematic_records = file.readlines()

    separated_data = parse_schematic_records(schematic_records)
        
    symbol_row_data = separated_data[1]
    number_row_data = separated_data[0]

    for item in number_row_data:
        print(str(item))
    
    sum_of_parts = evaluate_adjacency(symbol_row_data, number_row_data)
    print(sum_of_parts)
   
    end = time.perf_counter()
    print(f"\nExecution time {end - start:0.4f} seconds")
# Entrypoint
debug = False
main()
```
